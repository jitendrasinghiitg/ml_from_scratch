## Choice of Cost Function for Linear Regression

After undersatnding Linear Regression ,a question comes into mind why do we use MSE as cost function.

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### we can not use x, x^3,x^5 as our cost function as they do not have any minima

but absolute of these can be used ,so the candidate functions for our cost function are :

$abs(x), x^2, abs(x^3), x^4 ...$

```python
x = np.arange(0,2.025,0.025)

X_square = np.square(x)
X_cube = np.abs(x*x*x)
X_square_square = x*x*x*x

x_abs = np.abs(x)
```

```python
plt.figure(figsize=(10,5))

plt.plot(x,x_abs)
plt.plot(x,X_square)
plt.plot(x,X_cube)
plt.plot(x,X_square_square)
plt.xlabel("deviation of predicted value from actual value")
plt.ylabel("cost due to one single point")

plt.legend(('abs(X)','X^2', 'abs(X^3)','x^4'))
```

We can see when deviation is more than 1 ,$x^4$ is way more than abs(x) which means if we use $x^4$ as loss function, our loss funtion will be affected heavily by points with deviation more than 1.

and for deviation less than 1 ,x^4 is very less than other values ,which says loss due to near points will be least.


so effect of outliers on algo ,for various costfunction will be in order:

$abs(x) < x^2 <abs(x^3) <x^4 $ .. 

#### we understood that $abs(x^3) ,x^4 $ can not be used as cost function, as affect of outliers will be huge on model$

#### Now we are Left with $abs(x) , x^2$ 

```python
# lets just try to plot abs(x) and x^2 and there derivative
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline 

x = np.linspace(-5,5,100)

# the function, which is y = x^2 here
y = x**2
y_derivative = 2*x

# setting the axes at the centre
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1, 2, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the function
plt.plot(x,y, 'r')

ax = fig.add_subplot(1, 2, 2)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x,y_derivative, 'r')



# show the plot
plt.show()
```

If we use abs(x), derivative of it is either -1 or 1, For a point we never know how far we are from The minima 
ie derivative of abs(x) for x=2 and x=4 is 1 so just by looking at derivative we can no guess how far this point is from minima

, whereas derivative of x^2 has gives info about how far the point is from minima, ie for x=2, and x=4 derivative of x^2 is 4 and 8 which me shows 2 is closer to minima

Also in case of abs(x) ,derivative will be near to zero when half of the point are on both side's of minima , that's why algorithm will tries to find a line/ hyperplane ,such that there are equal no of point on both sides of line/ hyperplane
