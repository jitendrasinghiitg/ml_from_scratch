## Choice of Cost Function for Linear Regression

After undersatnding Linear Regression ,a question comes into mind why do we use MSE as cost function.

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### we can not use x, x^3,x^5 as our cost function as they do not have any minima

but absolute of these can be used ,so the candidate functions for our cost function are :

$abs(x), x^2, abs(x^3), x^4 ...$

```python
x = np.arange(0,2.025,0.025)

X_square = np.square(x)
X_cube = np.abs(x*x*x)
X_square_square = x*x*x*x

x_abs = np.abs(x)
```

```python
plt.figure(figsize=(10,5))

plt.plot(x,x_abs)
plt.plot(x,X_square)
plt.plot(x,X_cube)
plt.plot(x,X_square_square)
plt.xlabel("deviation of predicted value from actual value")
plt.ylabel("cost due to one single point")

plt.legend(('abs(X)','X^2', 'abs(X^3)','x^4'))
```

We can see when deviation is more than 1 ,$x^4$ is way more than abs(x) which means if we use $x^4$ as loss function, our loss funtion will be affected heavily by points with deviation more than 1.

and for deviation less than 1 ,x^4 is very less than other values ,which says loss due to near points will be least.


so effect of outliers on algo ,for various costfunction will be in order:

$abs(x) < x^2 <abs(x^3) <x^4 $ .. 

#### we understood that $abs(x^3) ,x^4 $ can not be used as cost function, as affect of outliers will be huge on model$

Gradient descent minimizes the first-order approximation (plus a quadratic function) to a function in each iteration. The first-order approximation to f at iteration k+1 is made at the solution obtained at iteration k.

$f(x) ~ f_k(x) := f(x_k) + <grad_f(x_k), x - x_k> + L/2||x - x_k||_2^2$

Now, $x_{k+1}$ is obtained by minimizing $f_k(x)$
Note that if it is an "unconstrained" minimization, the minimizer of $f_k(x)$ is given by $x_{k+1} = x_k - 1/L grad_f(x_k)$.
